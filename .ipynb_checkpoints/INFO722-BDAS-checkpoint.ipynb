03-DP

```python
# 3.1 Select the data

df = df[df['Country'].isin(['Australia', 'New Zealand'])]

years = [str(year) for year in range(1961, 2023)]

df_filtered = df[['Country'] + years]
```

```python
print(df_filtered)
```

```python
# 3.2 Clean the data

df_filtered.fillna(df_filtered.mean(), inplace=True)
```

```python
print(df_filtered)
```

```python
# 3.3 Construct the data

df_filtered.set_index('Country', inplace=True)

anz_average = df_filtered.mean(level=0).mean(axis=0)

anz_average = anz_average.reset_index()

anz_average.columns = ['Year', 'ANZ_Temperature_Change']
```

```python
print(anz_average)
```

```python
# 3.4  Integrate various data sources

df_2023 = pd.read_csv('merge.csv')

df_combined = pd.concat([anz_average, df_2023], ignore_index=True)

df_combined = df_combined.iloc[:, :-1]
```

```python
print(df_2023)
```

```python
print(df_combined)
```

```python
# 3.5 Format the data as required

df_combined['Year'] = pd.to_datetime(df_combined['Year'], format='%Y')
```

```python
print(df_combined)
```

```python
# 4.1 Reduce the data

# df = df[df['Country'].isin(['Australia', 'New Zealand'])]

# years = [str(year) for year in range(1961, 2023)]

# df_filtered = df[['Country'] + years]
```

```python
print(df_combined)
```

```python
print(df_filtered)
```

```python
# 4.2 Project the data

# df_filtered.set_index('Country', inplace=True)

# anz_average = df_filtered.mean(level=0).mean(axis=0)

# anz_average = anz_average.reset_index()

# anz_average.columns = ['Year', 'ANZ_Temperature_Change']

df_combined['Year'] = df_combined['Year'].dt.year

df_combined['Decade'] = (df_combined['Year'] // 10) * 10

decade_means = df_combined.groupby('Decade')['ANZ_Temperature_Change'].mean().reset_index()

decade_means.columns = ['Decade', 'Mean_ANZ_Temperature_Change']
```

```python
print(decade_means)
```

```python
# Show the decade data plot

plt.figure(figsize=(10, 5))

plt.plot(decade_means['Decade'], decade_means['Mean_ANZ_Temperature_Change'], marker='o')  

plt.title('Average Temperature Change by Decade for Australia and New Zealand')

plt.xlabel('Decade')

plt.ylabel('Mean Temperature Change (°C)')

plt.grid(True)

plt.show()
```

```python
# Show the yearly data plot

plt.figure(figsize=(12, 6)) 

plt.plot(df_combined['Year'], df_combined['ANZ_Temperature_Change'], marker='o', linestyle='-', color='b')

plt.title('Yearly Temperature Change for Australia and New Zealand')

plt.xlabel('Year')

plt.ylabel('Temperature Change (°C)')

plt.grid(True)

plt.xticks(df_combined['Year'], rotation=45)

plt.tight_layout() 

plt.show()
```

```python
# 6.3 Build/Select appropriate models and choose relevant parameters
# Convert these Pandas DataFrames to PySpark DataFrames

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("TemperatureChangeForecast").getOrCreate()

df_spark = spark.createDataFrame(df_combined)

from pyspark.ml.feature import VectorAssembler

from pyspark.ml.regression import LinearRegression

vectorAssembler = VectorAssembler(inputCols=['Year'], outputCol='features')

df_ml = vectorAssembler.transform(df_spark)
```

```python
df_ml.show()
```

```python
# Define the Linear Regression model

lr = LinearRegression(featuresCol='features', labelCol='ANZ_Temperature_Change')
```

07-DM

```python
# Split the data into training and testing sets

train_data, test_data = df_ml.randomSplit([0.8, 0.2])
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
# Fit the model
lr_model = lr.fit(train_data)

# Predictions
predictions = lr_model.transform(test_data)
```

```python
predictions.select("prediction", "ANZ_Temperature_Change", "features").show()
```

08-INT

```python
plt.figure(figsize=(10, 5))

plt.plot(predictions.toPandas()['Year'], predictions.toPandas()['prediction'], label='Predicted')

plt.plot(predictions.toPandas()['Year'], predictions.toPandas()['ANZ_Temperature_Change'], label='Actual')

plt.title('Comparison of Predicted and Actual Temperature Changes')

plt.xlabel('Year')

plt.ylabel('Temperature Change (°C)')

plt.legend()

plt.show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(labelCol="ANZ_Temperature_Change", predictionCol="prediction", metricName="rmse")

rmse = evaluator.evaluate(predictions)

print("Root Mean Squared Error (RMSE) on test data =", rmse)
```

```python
from pyspark.ml.regression import LinearRegression

# Adjusting parameters for regularization
lr = LinearRegression(featuresCol='features', labelCol='ANZ_Temperature_Change', 
                      regParam=0.1, elasticNetParam=0, maxIter=100)

# Fit the model
lr_model = lr.fit(train_data)

# Evaluate the model
predictions = lr_model.transform(test_data)

evaluator = RegressionEvaluator(labelCol="ANZ_Temperature_Change", predictionCol="prediction", metricName="rmse")

rmse = evaluator.evaluate(predictions)

print("Root Mean Squared Error (RMSE) on test data =", rmse)
```

```python
from pyspark.ml.regression import DecisionTreeRegressor

# Define the Decision Tree model
dt = DecisionTreeRegressor(featuresCol='features', labelCol='ANZ_Temperature_Change')

# Fit the model
dt_model = dt.fit(train_data)

# Predictions
predictions = dt_model.transform(test_data)

# Evaluate the model
evaluator = RegressionEvaluator(labelCol="ANZ_Temperature_Change", predictionCol="prediction", metricName="rmse")

rmse = evaluator.evaluate(predictions)

print("Root Mean Squared Error (RMSE) on test data =", rmse)
```
